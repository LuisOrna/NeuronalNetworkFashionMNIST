##1.2 One-Hot Encoding

```python
#Creating a funtion to do the one hot encoding
def one_hot_encode(y, num_classes=10):

  #Creating a matrix with 0
  matrix = np.zeros(shape=(y.shape[0], num_classes))

  #Putting 1 in the right position
  matrix[np.arange(stop=y.shape[0]), y] = 1

  return matrix
```

#4. BACKPROPAGATION

```python
# Gradient of output layer (Softmax + Cross-Entropy)
dZ2 = A2 - y_train_encoded  # (60000, 10)

# Gradients for W2 and b2
dW2 = (1/60000) * np.dot(A1.T, dZ2)  # (10, 10)
db2 = (1/60000) * np.sum(dZ2, axis=0, keepdims=True)  # (1, 10)

# Gradient propagated to hidden layer
dA1 = np.dot(dZ2, W2.T)  # (60000, 10)

# Gradient of ReLU (derivative is 1 if z>0, else 0)
dZ1 = dA1 * (Z1 > 0)  # (60000, 10)

# Gradients for W1 and b1
dW1 = (1/60000) * np.dot(X_train_nor_flat.T, dZ1)  # (784, 10)
db1 = (1/60000) * np.sum(dZ1, axis=0, keepdims=True)  # (1, 10)
```

```python
print("dW1:", dW1.shape)
print("db1:", db1.shape)
print("dW2:", dW2.shape)
print("db2:", db2.shape)
```

```python
learning_rate = 0.1

W1 = W1 - learning_rate * dW1
b1 = b1 - learning_rate * db1
W2 = W2 - learning_rate * dW2
b2 = b2 - learning_rate * db2
```

#5. Neuronal Network 1

This has three layers:


*   Entry layer with 784 neurons.

*   Hidden layer with 10 layers.
*   Exit layer with 10 layers.



```python
# Forward pass (AGAIN)
Z1 = np.dot(X_train_nor_flat, W1) + b1
A1 = np.maximum(0, Z1)
Z2 = np.dot(A1, W2) + b2
A2 = np.exp(Z2) / np.sum(np.exp(Z2), axis=1, keepdims=True)

# Loss y accuracy
loss = -np.mean(np.sum(y_train_encoded * np.log(A2 + 1e-8), axis=1))
predictions = np.argmax(A2, axis=1)
accuracy = np.mean(predictions == y_train)

print("Loss:", loss)
print("Accuracy:", accuracy)
```

```python
#Irering multiple times
epochs = 101

#Starting time
start_time = time.time()

for epoch in range(epochs):
    # Forward pass
    Z1 = np.dot(X_train_nor_flat, W1) + b1
    A1 = np.maximum(0, Z1)
    Z2 = np.dot(A1, W2) + b2
    A2 = np.exp(Z2) / np.sum(np.exp(Z2), axis=1, keepdims=True)

    # Loss
    loss = -np.mean(np.sum(y_train_encoded * np.log(A2 + 1e-8), axis=1))

    # Backpropagation
    dZ2 = A2 - y_train_encoded
    dW2 = (1/60000) * np.dot(A1.T, dZ2)
    db2 = (1/60000) * np.sum(dZ2, axis=0, keepdims=True)
    dA1 = np.dot(dZ2, W2.T)
    dZ1 = dA1 * (Z1 > 0)
    dW1 = (1/60000) * np.dot(X_train_nor_flat.T, dZ1)
    db1 = (1/60000) * np.sum(dZ1, axis=0, keepdims=True)

    # Update weights
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    # Print every 10 epochs
    if epoch % 10 == 0:
        predictions = np.argmax(A2, axis=1)
        accuracy = np.mean(predictions == y_train)
        print(f"Epoch {epoch}: Loss = {loss:.4f}, Accuracy = {accuracy:.4f}")


training_time = time.time() - start_time
```

```python
# Forward pass con test set
Z1_test = np.dot(X_test_nor_flat, W1) + b1
A1_test = np.maximum(0, Z1_test)
Z2_test = np.dot(A1_test, W2) + b2
A2_test = np.exp(Z2_test) / np.sum(np.exp(Z2_test), axis=1, keepdims=True)

# Accuracy en test
predictions_test = np.argmax(A2_test, axis=1)
accuracy_test = np.mean(predictions_test == y_test)

print(f"Test Accuracy: {accuracy_test:.4f}")
```

```python
#Saving results
results = []

#Funtion to save
def save_results(try_number, hidden_neurons, hidden_layers, epoch, accuracy, accuracy_test, loss, time_to_train):
  network = {'try': try_number,
             'hidden_neurons': hidden_neurons,
             'hidden_layers': hidden_layers,
             'epoch': epoch,
             'accuracy': accuracy,
             'accuracy_test': accuracy_test,
             'loss': loss,
             'time': time_to_train
             }
  results.append(network)
  return results
```

#6. Refactoring

I have decided to refactor all code in funtions in order to make several tries and arquitectures easily.

```python
#Bulding parameters
def initialize_parameters( hidden_size, imput_size=784, output_size=10, scale_factor=0.01):
  '''Initialize main parameters of neuronal network, returns W1, b1 W2, b2'''
  np.random.seed(42)

  W1= np.random.randn(imput_size, hidden_size) * scale_factor
  b1 = np.zeros(shape=(1, hidden_size))
  W2 = np.random.randn(hidden_size, output_size) * scale_factor
  b2 = np.zeros(shape=(1, output_size))

  return W1, b1, W2, b2

```

```python
#Forward Pass
def forward_pass(X, W1, b1, W2, b2):
  '''compute forward propagation'''

  #Layer 1
  Z1 = np.dot(a=X, b=W1) + b1
  A1 = np.maximum(0, Z1)

  #Layer 2
  Z2 = np.dot(a=A1, b=W2) + b2
  #softmax
  exp_z2 = np.exp(Z2)
  A2 = exp_z2 / np.sum(exp_z2, axis=1, keepdims=True)

  return Z1, A1, Z2, A2
```

```python
#Estimating Loss
def compute_loss(A2, y_encoded):
  '''Compute cross-entropy loss, return loss'''
  loss = -np.mean(np.sum(y_encoded * np.log(A2 + 1e-15), axis=1))

  return loss
```

```python
#Backpropagation
def backward_pass(X, y_encoded, Z1, A1, A2, W2):
    '''Compute gradients using backpropagation'''
    m = X.shape[0]  # Number of samples

    # Output layer gradients
    dZ2 = A2 - y_encoded
    dW2 = (1/m) * np.dot(A1.T, dZ2)
    db2 = (1/m) * np.sum(dZ2, axis=0, keepdims=True)

    # Hidden layer gradients
    dA1 = np.dot(dZ2, W2.T)
    dZ1 = dA1 * (Z1 > 0)  # ReLU derivative
    dW1 = (1/m) * np.dot(X.T, dZ1)
    db1 = (1/m) * np.sum(dZ1, axis=0, keepdims=True)

    return dW1, db1, dW2, db2
```

```python
#Training
def train_loop(X, y_encoded, W1, b1, W2, b2, epochs, learning_rate):
    '''Execute training loop with given parameters'''

    #time
    start_time = time.time()

    for epoch in range(epochs):
        # Forward pass
        Z1, A1, Z2, A2 = forward_pass(X, W1, b1, W2, b2)

        # Compute loss
        loss = compute_loss(A2, y_encoded)

        # Backward pass
        dW1, db1, dW2, db2 = backward_pass(X, y_encoded, Z1, A1, A2, W2)

        # Update parameters
        W1 -= learning_rate * dW1
        b1 -= learning_rate * db1
        W2 -= learning_rate * dW2
        b2 -= learning_rate * db2

        # Print progress
        if epoch % 10 == 0:
            predictions = np.argmax(A2, axis=1)
            accuracy = np.mean(predictions == np.argmax(y_encoded, axis=1))
            print(f"Epoch {epoch}: Loss = {loss:.4f}, Accuracy = {accuracy:.4f}")

    #Computing time
    training_time = time.time() - start_time

    return W1, b1, W2, b2, loss, accuracy, training_time
```

```python
#Evaluating
def evaluate(X, y, W1, b1, W2, b2):
    '''Evaluate model on given dataset'''
    # Forward pass
    _, _, _, A2 = forward_pass(X, W1, b1, W2, b2)

    # Compute accuracy
    predictions = np.argmax(A2, axis=1)
    accuracy = np.mean(predictions == y)

    return accuracy
```
